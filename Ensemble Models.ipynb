```python
import pandas as pd 
data = pd.read_csv('churn_prediction_simple.csv')
data.info()
```

```python
# seperating dependent and independent variables 
X = data.drop(columns = ['churn', 'customer_id'])
Y = data['churn']
```

```python
#Scaling the dataset 
from sklearn.preprocessing import StandardScaler 
scaler = StandardScaler()
scaled_X = scaler.fit_transform(X)
```

```python
#splitting the dataset 
from sklearn.model_selection import train_test_split as tts 
x_train, x_test, y_train, y_test = tts(scaled_X, Y, train_size = 0.80, stratify= Y )
x_train.shape, x_test.shape, y_train.shape , y_test.shape
```

# Model Building, Bagging Logistic Regrassion

```python
from sklearn.ensemble import BaggingClassifier as BC 
Classifier = BC()
Classifier.fit(x_train, y_train)
Classifier.get_params()
```

```python
from sklearn.linear_model import LogisticRegression as LR 
Classifier= BC(base_estimator = LR(),
               n_estimators = 150,
               n_jobs = -1,
               random_state = 42)
Classifier.fit(x_train, y_train)
predicted_values = Classifier.predict(x_train)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_train, predicted_values))
```

```python
predicted_values = Classifier.predict(x_test)
print(classification_report(y_test, predicted_values))
```

```python
from sklearn.metrics import f1_score
def calc_score(model, x1, y1, x2, y2):
    model.fit(x1,y1)
    
    predict = model.predict(x1)
    f1 = f1_score(y1, predict)
    
    predict = model.predict(x2)
    f2 = f1_score(y2, predict)
    
    return f1, f2
```

```python
estimators = [i for i in range(1,600,10)]
train= []
test= []

for i in estimators:
    model = RFC(class_weight = 'balanced_subsample', 
                                n_estimators= i,
                                n_jobs = -1,
                                max_depth = 7,
                                random_state=101)
    f1,f2= calc_score(model, x_train, y_train, x_test, y_test)
    train.append(f1)
    test.append(f2)
```

```python
effect(train, test, range(1,600,10), 'n_estimators')
```

# max_samples

```python
maxsamples = [i/1000 for i in range(1,1000)]
train= []
test= []

for i in maxsamples:
    model = RFC(class_weight = 'balanced_subsample', 
                                n_estimators= 50,
                                n_jobs = -1,
                                max_depth = 7,
                                random_state=101, max_samples= i)
    f1,f2= calc_score(model, x_train, y_train, x_test, y_test)
    train.append(f1)
    test.append(f2)
```

```python
effect(train, test, maxsamples, 'bootstrap sample fraction ')
```

```python
maxfeatures = range(1,X.shape[1])
train= []
test= []

for i in maxfeatures:
    model = RFC(class_weight = 'balanced_subsample', 
                                n_estimators= 50,
                                n_jobs = -1,
                                max_depth = 7,
                                random_state=101, max_features= i)
    f1,f2= calc_score(model, x_train, y_train, x_test, y_test)
    train.append(f1)
    test.append(f2)
```
